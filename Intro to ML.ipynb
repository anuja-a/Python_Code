```python
import pandas as pd
import numpy as np
#HI
```

```python
x1=np.random.randint(low=1,high=20,size=(50,))


x2=np.random.randint(low=1,high=20,size=(50,))
```

```python
y=3+2*x1-4*x2+np.random.random((50,))
```

```python


x=pd.DataFrame({'intercept':np.ones_like(x1),'x1':x1,'x2':x2})

w=np.zeros(x.shape[1]) #weights or betas
print(x)
print(x.shape)
print(w.shape)
```

```python
#calculating predicted y
#beta *x
def myprediction(features,weights):

    predictions=np.dot(features,weights)
    return(predictions)

myprediction(x,w)  # function call

#myprediction(x,w).shape
```

```python
# calculating error
#yact - ypred
#yact - beta *X
def myerror(target,features,weights):
    error=target-myprediction(features,weights)
    return(error)
```

```python
# Calculating cost i.e error^2
# ((e*e.T))

def mycost(target,features,weights):
    error=myerror(target,features,weights)
    cost=np.dot(error.T,error)     #gives single error value 
    return(cost)

mycost(y,x,w)
```

```python
# -(Error *X)
def gradient(target,features,weights):
    
    error=myerror(target,features,weights)
    
    gradient=-np.dot(features.T,error)
    
    return(gradient)

gradient(y,x,w)
```

```python
#pass various values of lr and check
#Gradient of beta_0 is small. Hence for small lr, change is small
def my_lr(target,features,learning_rate):
    
    weights=np.zeros(features.shape[1])
    
    for i in np.arange(100000):
        weights -= learning_rate*gradient(target,features,weights)
        weights[0] -=10*learning_rate*gradient(target,features,weights)[0] # if you comment this out , your coefficient estimates will not be as good as sklearn, why?
        if i%100==0:
            print(mycost(target,features,weights),weights)
    return(weights)

my_lr(y,x,0.000001)
```
